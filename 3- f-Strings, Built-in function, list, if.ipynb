# Indexing



It is a way to access individual elements in a list (or any sequence like strings or tuples). Each element in the list is assigned an index, which is its position in the list, starting from 0.

**Positive indexing** starts from 0 for the first item, 1 for the second item, and so on.

**Negative indexing** starts from -1 for the last item, -2 for the second last, and so on.

# Slicing






- Slicing is a technique used to extract a portion of a list (or any sequence like strings or tuples) by specifying a range of indices.

- Slicing creates a new list containing the specified elements.

- list[start:end:step]

1. start: The index to begin the slice.
2. end: The index to end the slice.
3. step: The interval between elements.


```python
fruits = ["apple", "banana", "cherry", "orange", "grape"]
# Reverse slicing
print(fruits[::-1])
```

```python
fruits = ["apple", "banana", "cherry", "orange", "grape"]

# Basic slicing
print(fruits[1:4])

# Slicing with a step
print(fruits[0:5:2])

# Omitting start or end
print(fruits[:3])
print(fruits[2:])

```
